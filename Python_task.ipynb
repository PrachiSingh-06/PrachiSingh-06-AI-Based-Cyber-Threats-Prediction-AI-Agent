Write a Python program that takes a string input and checks if it is a palindrome, reverses it, and displays whether it reads the same backward or not.

```python
# Static input (you can change the string value here)
text = "Madam"

# Convert the string to lowercase so that case differences don't matter
lower_text = text.lower()

# Reverse the string using slicing
reversed_text = lower_text[::-1]

# Display the reversed string
print("Reversed string:", reversed_text)

# Compare the original (lowercase) string with its reversed version
if lower_text == reversed_text:
    print("The given string is a palindrome.")
else:
    print("The given string is not a palindrome.")

# Display the original string again for clarity
print("Original string was:", text)

```

Create a dictionary of employee names and their salaries. Then: Add a new employee, find the employee with the highest salary, and convert the dictionary into a Pandas DataFrame.

```python
import pandas as pd

# Step 1: Create a dictionary of employee names and their salaries
employees = {
    "Rahul": 50000,
    "Sneha": 60000,
    "Arjun": 55000,
    "Priya": 65000
}

# Step 2: Add a new employee to the dictionary
employees["Vikas"] = 70000

# Step 3: Find the employee with the highest salary
topper_name = max(employees, key=employees.get)
topper_salary = employees[topper_name]

# Step 4: Convert the dictionary into a Pandas DataFrame
df = pd.DataFrame(list(employees.items()), columns=["Employee Name", "Salary"])

# Step 5: Display the DataFrame and topper details
print("Employee Data:\n")
print(df)
print("\nTopper Employee:", topper_name)
print("Highest Salary:", topper_salary)
```

Create a list of 10 random numbers. Then: Find the sum and average of the numbers, remove duplicates, and display the list in ascending order.

```python
import random

# Step 1: Create a list of 10 random numbers between 1 and 100
numbers = [random.randint(1, 100) for _ in range(10)]

# Step 2: Display the original list
print("Original List:", numbers)

# Step 3: Find the sum of all numbers
total_sum = sum(numbers)

# Step 4: Find the average of the numbers
average = total_sum / len(numbers)

# Step 5: Remove duplicate values using set() and convert back to list
unique_numbers = list(set(numbers))

# Step 6: Sort the list in ascending order
unique_numbers.sort()

# Step 7: Display all results
print("\n--- Results ---")
print("Sum of numbers:", total_sum)
print("Average of numbers:", average)
print("List after removing duplicates:", unique_numbers)
print("Sorted List (Ascending):", unique_numbers)
```

Create a tuple of 10 city names (with some duplicates). Then: Remove the duplicate cities, convert the tuple to a list, and sort the list alphabetically.


```python
# Step 1: Create a tuple of city names with some duplicates
cities = ("Delhi", "Mumbai", "Chennai", "Kolkata", "Delhi", "Pune", "Jaipur", "Mumbai", "Surat", "Bhopal")

# Step 2: Display the original tuple
print("Original Tuple:", cities)

# Step 3: Remove duplicate city names by converting the tuple into a set
unique_cities = set(cities)

# Step 4: Convert the set back into a list
city_list = list(unique_cities)

# Step 5: Sort the list alphabetically
city_list.sort()

# Step 6: Display the final unique and sorted city list
print("\nUnique and Sorted City List:", city_list)
```

Generate a 4Ã—4 matrix of random integers using NumPy. Then: Find the sum of each row and column, and display the matrix using a heatmap with seaborn.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Generate a 4x4 matrix of random integers between 1 and 100
matrix = np.random.randint(1, 101, (4, 4))

# Step 2: Display the generated matrix
print("Generated 4x4 Matrix:\n", matrix)

# Step 3: Calculate the sum of each row
row_sum = np.sum(matrix, axis=1)

# Step 4: Calculate the sum of each column
col_sum = np.sum(matrix, axis=0)

# Step 5: Display the results
print("\nSum of each row:", row_sum)
print("Sum of each column:", col_sum)

# Step 6: Plot the matrix using a heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(matrix, annot=True, cmap="coolwarm", fmt="d")
plt.title("4x4 Random Integer Matrix Heatmap")
plt.xlabel("Columns")
plt.ylabel("Rows")
plt.show()
```

Plot a pie chart of daily time spent on activities using Matplotlib. Include: labels, percentage display, and a title for the chart.


```python
import matplotlib.pyplot as plt

# Step 1: Data for daily activities (in hours)
activities = ['Sleep', 'Work', 'Study', 'Exercise', 'Leisure', 'Others']
hours = [8, 7, 3, 1, 3, 2]

# Step 2: Choose custom colors for each activity
colors = ['#FF9999', '#66B2FF', '#99FF99', '#FFCC99', '#FFD700', '#DDA0DD']

# Step 3: Plot the pie chart
plt.figure(figsize=(7, 7))
plt.pie(hours,
        labels=activities,
        autopct='%1.1f%%',   # Display percentages with 1 decimal
        startangle=90,       # Rotate chart for better view
        colors=colors,
        shadow=True)

# Step 4: Add a title
plt.title('Daily Time Spent on Activities', fontsize=14)

# Step 5: Display the pie chart
plt.show()
```

Write a Python program to create employee data with Name, Department, and Salary. Filter salaries above 50,000, group by department, find the average salary, and plot it using a colored bar chart.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Creating dummy employee data
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace', 'Hannah', 'Ian', 'Jack'],
    'Department': ['HR', 'IT', 'Finance', 'IT', 'Finance', 'HR', 'IT', 'Finance', 'HR', 'IT'],
    'Salary': [48000, 60000, 52000, 75000, 49000, 51000, 80000, 55000, 47000, 72000]
}

# Converting dictionary to DataFrame
df = pd.DataFrame(data)

# Filtering employees with salary > 50,000
filtered_df = df[df['Salary'] > 50000]

# Grouping by department and counting employees
dept_count = filtered_df.groupby('Department')['Name'].count()

# Calculating average salary per department
dept_salary_avg = filtered_df.groupby('Department')['Salary'].mean()

# Display results
print("Filtered Employees (Salary > 50,000):\n", filtered_df)
print("\nEmployee Count by Department:\n", dept_count)
print("\nAverage Salary by Department:\n", dept_salary_avg)

# Plot department-wise average salary
dept_salary_avg.plot(kind='bar', color=['skyblue', 'lightgreen', 'orange'])
plt.title('Average Salary by Department (Salary > 50,000)')
plt.xlabel('Department')
plt.ylabel('Average Salary')
plt.show()

```

Create a class CyberThreatDetector with attributes: ip_address, threat_score, and timestamp. Include methods to evaluate if the threat is critical or safe and display the detection summary.

```python
import datetime

class CyberThreatDetector:
    def __init__(self, ip_address, threat_score, timestamp):
        self.ip_address = ip_address
        self.threat_score = threat_score
        self.timestamp = timestamp

    def evaluate_threat(self):
        if self.threat_score >= 80:
            return "Critical"
        elif self.threat_score >= 40:
            return "Moderate"
        else:
            return "Safe"

    def display_summary(self):
        threat_status = self.evaluate_threat()
        print("=== Cyber Threat Detection Summary ===")
        print(f"IP Address     : {self.ip_address}")
        print(f"Threat Score   : {self.threat_score}")
        print(f"Timestamp      : {self.timestamp}")
        print(f"Threat Status  : {threat_status}")
        print("======================================")

# Static input example
detector = CyberThreatDetector("192.168.1.5", 76, datetime.datetime.now())
detector.display_summary()
```
